```python
import bayesopt

import matplotlib.pyplot as plt

from jax.config import config; config.update("jax_enable_x64", True)
from functools import partial
import jax.numpy as np
import jax

import GPyOpt
```

```python
def target(x):
    return x*np.sin(x)

    #(x1-2.)**2 + 10*(x2+3)**2

# def target2(x):
#     return target(x[0],x[1])
#     #return x*np.sin(x)
#     #return -np.exp(-(x - 3)**2) - np.exp(-(x - 6)**2/10) - 1/ (x**2 + 1)
```

```python
KAPPA = 5
x = np.linspace(0, 20, 1000)
y = target(x)
plt.plot(x, y)
```

```python
# %%timeit
BO = bayesopt.BayesOpt(f=target,
              initial_input=np.array([0.]),
              alpha=0.1,
              kernel=bayesopt.kernel.GaussianRBFKernel(a=1,h=1),
              acq=jax.jit(partial(bayesopt.acquisition.LCB,kappa=10.)),
              acq_optim=partial(bayesopt.acquisition_optimizer.AcquisitionGridOptimizer,
                               bounds=np.array([0,20]),step=0.2))
BO.run_optim(50)
```

```python
bounds = [{'name': 'x1', 'type': 'continuous', 'domain': (0,20)}]
myBopt = GPyOpt.methods.BayesianOptimization(f=target, 
                                             domain=bounds,acquisition_type='LCB',initial_design_numdata=1)
myBopt.run_optimization(max_iter=50)
```

```python
plt.plot(BO.value_history+18,'o',label='MyBO')
plt.plot([np.min(BO.value_history[:i+1]+18) for i in range(50)],label='MyBO')

plt.plot(myBopt.Y+18,'x',label='GPyOpt')
plt.plot([np.min(myBopt.Y[:i+1]+18) for i in range(50)],label='GPyOpt')
plt.legend()
# plt.xlim(0,40)
plt.yscale('log')
```
