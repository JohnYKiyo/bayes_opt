```python
import time
class BayesOpt(object):
    def __init__(self, f, initial_X, acq, acq_optim,*args, **kwargs):
        self.__objectivefunction = f
        self.__initial_X = initial_X
        self.__initial_Y = f(initial_X)
        self.__best_params = None
        self.__best_value = None
        
        ##init GPR
        self.__kernel = kwargs.pop('kernel',bayesopt.kernel.GaussianRBFKernel(h=1.0,a=1.0))
        self.__alpha = kwargs.pop('alpha',1e-6)
        self.__gpr = bayesopt.GPR(X_train=self.__initial_X,
                                  Y_train=self.__initial_Y,
                                  alpha=self.__alpha,
                                  kernel=self.__kernel)
        
        self.__X_history = self.__gpr.X_train
        self.__Y_history = self.__gpr.Y_train
        
        ##init acquuisition function
        self.__acq = acq
        self.__acq_optimizer = acq_optim
        self.__acq_history = []
    def suggest_next_locations(self):
        predict_val, next_location, acqval = self.__acq_optimizer(self.__gpr, self.__acq)
        self.__acq_history.append(acqval.ravel())
        return next_location
    
    def run_optim(self,n_iter):
        for i in range(n_iter):
            next_loc = self.suggest_next_locations()
            Y_obs = self.__objectivefunction(next_loc)
            self.__gpr.append_data(next_loc, Y_obs)
            self.__X_history = self.__gpr.X_train
            self.__Y_history = self.__gpr.Y_train
            
        self.__best_value = np.max(self.__Y_history)
        self.__best_params = self.__X_history[np.argmax(self.__Y_history)]

class GridSampler(object):
    
    def __init__(self, Xmin, Xmax, step):
        self.__Xmin = bayesopt.utils.transform_data(Xmin)
        self.__Xmax = bayesopt.utils.transform_data(Xmax)
        ##data dimention check
        if self.__Xmin.shape != self.__Xmax.shape :
            raise ValueError('Xmin,Xmax should be same size.')
        self.__ndim = len(self.__Xmin)
        
        ##step size init
        self.__step = bayesopt.utils.transform_data(step)
        if (self.__step.shape != (self.__ndim,1)):
            if self.__step.shape[1] != 1:
                raise ValueError('step should be an 1-D array_like or a numerical value.')
            if self.__step.shape[0] == 1:
                self.__step = np.full_like(self.__Xmin,step)
            else:
                raise ValueError(f'step shape should be same shape of Xmin and Xmax: {self.__Xmin.shape},\
                                 but get{self.__step.shape}')
        
        ##generate grid points
        d_list = tuple(np.arange(mi,ma,st) for mi,ma,st in zip(self.__Xmin,self.__Xmax,self.__step))
        self.grid = np.array(np.meshgrid(*d_list)).reshape(self.__ndim,-1).T
    
        ###iterator###
        self.__i = 0
        
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.__i == len(self.grid):
            raise StopIteration()
        ret = tuple(self.grid[self.__i])
        self.__i += 1
        return ret
    
    def __call__(self):
        return self.grid

def AcquisitionGridOptimizer(gpr, acq, Xmin, Xmax, step):
    GS = GridSampler(Xmin,Xmax,step)
    mu_s, std_s = gpr.posterior_predictive(GS.grid,return_std=True)
    val = acq(mu_s, std_s)
    return np.min(val), GS.grid[np.argmin(val)],val
```

```python
%%timeit
BO = BayesOpt(f=f,
             initial_X=0., 
             acq=bayesopt.acquisition.UCB,
             acq_optim=partial(AcquisitionGridOptimizer,Xmin=0,Xmax=10,step=0.01))
             #(self, f, initial_X, acq, acq_optim,*args, **kwargs):
BO.run_optim(20)
```
